```python

#print step by step by color

#sort stud count by largest descending
#for each step, display an image showing the pixels for that color and the number of pixels between a corner of the object and the edge?
#color-by-color, use BFS to identify "clumps" and print with reference to distance in a straight cardinal direction to closest edge or existing clump?

#mergesort for stud_count!

def mergesort_index_count(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        left = arr[:mid]
        right = arr[mid:]
        
        mergesort(left)
        mergesort(right)
        i=j=k
        
        while(i<len(left) and j<len(right)):
            if left[i][1] > right[j][1]:
                arr[k]=left[i]
                i+=1
            else:
                arr[k]=right[j]
                j+=1
            k+=1
            
        while (i<len(left)):
            arr[k]=left[i]
            i+=1
            k+=1
        while (j<len(right)):
            arr[k]=right[j]
            j+=1
            k+=1


def sort_part_count(color_index_count,color_count):
    #create new list
    unsorted_list = np.array([])
    for index in np.arange(color_count):
        if color_index_count[index] != 0:
            np.append(unsorted_list,[index,color_index_count[index]])
    mergesort(unsorted_list)
    return unsorted_list

#functions for going through the sorted list in order of most studs to least and displaying image by image
import collections
#part_table = array that represents the color index for each pixel
#sorted_color_index_count = list of (index,count) sorted in descending count order
def build_instructions(part_table, sorted_color_index_count,height,width,color_table):
    curr_color_iter = 0
    images = np.array()
    partial_part_table = np.full((height,width),[-1,0])
    for color in sorted_color_index_count:
        loc_list = np.array()
        for row in np.arange(height):
            for column in np.arange(width):
                if part_table[row][column] == color[0]:
                    partial_part_table[row][column][0] = color[0]
                    np.append(loc_list,[row,column])
        imgstep = build_instruction_step(partial_part_table, sorted_color_index_count[curr_color_iter:], loc_list, color[0],color_table)
        curr_color_iter+=1
        np.append(images,imgstep)
    images[0].save(r'instructions.pdf',save_all=True,append_images=images[1:])


from PIL import ImageFont
from PIL import ImageDraw
#build_instruction_step takes the part_table as currently processed and runs BFS to identify clusters of pixels
#partial_part_table: the partially built array signifying which color index corresponds to each pixel
    #initially all -1; as the table is built, each color is added sequentially by pixel count
#remaining_index_count: the list of color indexes sorted by pixel count, with only the colors not already processed
#pixel_list: list of pixels corresponding to the color being processed
#index: index of the color being processed
def build_instruction_step(partial_part_table, remaining_index_count, pixel_list, index, color_table):
    #build an image array that is height*partsize x width*partsize with default colors, 
        #where partsize is the square size of the part in question 
    #populate image array with the colors from the partial_part_table (before the index to be built)
    #for each cluster,find closest ref, add black outlines to each part in the cluster, add a straight line down the middle
        #from the cluster pixel to the reference pixel and add text along that line to signify how far way it is (unless it is edge)
    shape = partial_part_table.shape
    partsize = 15
    img = np.array((shape[0]*partsize,shape[1]*partsize),[238,238,238])
    for row in np.arange(shape[0]):
        for column in np.arange(shape[1]):
            if partial_part_table[row][column] != -1:
                fill_in_part(img, row, column, index, color_table, partsize)            
    clusters = np.array()
    cluster_iter = 1
    while pixel_list.size > 0:
        np.append(clusters, find_cluster(partial_part_table,remaining_index_count,pixel_list,index, cluster_iter))
    
    written_pixels = np.array()
    for cluster_pixels in clusters:
        #fill in all the cluster pixels with outlines
        for pixel in cluster_pixels:
            fill_in_part(img,pixel[0],pixel[1],index,color_table,partsize)
            fill_in_outline(img,pixel[0],pixel[1],index,partsize)
        cluster_refs = find_closest_references(cluster_list,partial_part_table,index)
        #draw line down the middle of pixels from cluster pixel to ref
        #draw text at halway point of each ref line
        deltax = cluster_refs[0][0]-cluster_refs[1][0]
        deltay = cluster_refs[0][1]-cluster_refs[1][1]
        deltax=abs(deltax)
        deltay=abs(deltay)
        last_loc = cluster_refs[0]
        if deltax != 0:
            #drawline, drawtext, update last_loc and add text location to written_pixels
            midpoint = fill_in_line(img,last_loc,0,deltax,partsize)
            last_loc[1]+=deltax
            ref_cont = str(deltax)
            loc = midpoint
            if loc in written_pixels:
                loc = [midpoint[0]+1,midpoint[1]+1]
            np.append(written_pixels,loc)
            ref_text(img,loc,ref_cont)
            
        if deltay!=0:
            #drawline from last_loc, drawtext and add to written_pixels
            midpoint = fill_in_line(img,last_loc,1,deltay,partsize)
            ref_cont = str(deltay)
            loc = midpoint
            if loc in written_pixels:
                loc = [midpoint[0]+1,midpoint[1]+1]
            np.append(written_pixels,loc)
            ref_text(img,loc,ref_cont)
            
    #expand image up to show color name and number of parts; img.paste
    new_img = np.array((shape[0]*partsize+60,shape[1]*partsize),[238,238,238])
    new_img.paste(img,(60,0))
    ref_cont = str(pixel_list.size) + " pieces of " + color_name(index, color_table)
    ref_text(new_img, [15,15], ref_cont)
    return new_img

#takes image (the full expanded instruction image), the part location to be filled in (in terms of the not-as-yet expanded table)
# the index of the color to be used, the color_table to be used, and the size of the expanded part
def fill_in_part(image, row, col, index, color_table, partsize):
    img_start_row = row*partsize
    img_start_col = height*partsize
    color = color_table[index][3]
    for i in np.arange(img_start_row,img_start_row+partsize):
        for j in np.arange(img_start_col,img_start_col+partsize):
            image[i][j]=color

#returns midpoint of line for text drawing
def fill_in_outline(image, row, col, index, partsize):
    img_start_row = row*partsize
    img_start_col = col*partsize
    img_end_row = (row+1)*partsize - 1
    img_end_col = (col+1)*partsize - 1
    for i in np.arange(partsize):
        image[img_start_row][img_start_col + i] = [0,0,0]
        image[img_end_row][img_end_col - i] = [0,0,0]
        image[img_start_row + i][img_end_col] = [0,0,0]
        image[img_end_row - i][img_start_col] = [0,0,0]
        
def fill_in_line(image, start, x_or_y, delta, partsize):
    midpoint = [-1,-1]
    if x_or_y == 0:
        #iter thru cols from start to start+delta, where 3 pixels from center outwards are filled black
        #mark midpoint
        for i in np.arange(start[1],start[1]+delta):
            exp_col = i*partsize
            for d in np.arange(partsize):
                exp_col = d + exp_col 
                center = [start[0]*partsize + 7,exp_col]
                top = [start[0]*partsize + 6, exp_col]
                bottom = [start[0]*partsize + 8, exp_col]
                image[center[0]][center[1]] = [0,0,0]
                image[top[0]][top[1]] = [0,0,0]
                image[bottom[0]][bottom[1]] = [0,0,0]
        midpoint = [start[0], (start[1]+(delta/2))]

    else:
        #iter thru rows from start to start+delta as above
        #mark midpoint
        for i in np.arange(start[0],start[0]+delta):
            exp_row = i*partsize
            for d in np.arange(partsize):
                exp_col = d + exp_row 
                center = [exp_row,start[1]*partsize + 7]
                top = [exp_row, start[1]*partsize + 6]
                bottom = [exp_row, start[1]*partsize + 8]
                image[center[0]][center[1]] = [0,0,0]
                image[top[0]][top[1]] = [0,0,0]
                image[bottom[0]][bottom[1]] = [0,0,0]
        midpoint = [start[0]+(delta/2),start[1]]
    
    return midpoint
                    
def ref_text(img, pixel, text):
    font = ImageFont.truetype("arial.ttl",5)
    draw = ImageDraw.draw(img)
    location = (pixel[0],pixel[1])
    draw.text(location, text,font=font)
    

    #append '0' to every element in new table to signify that the pixel has not been "discovered": upon discovery,
        #mark it with a cluster label
    #run modified BFS on the first pixel in the list, with each "discovery" removing it from the list and adding it
            #to a list of clusters
        #rerun whenever there are still pixels in the list
        #save every root node as the start reference for a cluster
        #modified BFS: check surrounding pixels: if '0' AND contains current index, mark as the current cluster
    #take list of clusters and for each cluster, run find_closest_reference, which returns an image with references
    
    
#run BFS to identify a cluster
#as pixels are assigned to clusters, remove them from the remaining_pixel_list
def find_cluster(partial_part_table, remaining_index_count,remaining_pixel_list,index,curr_cluster):
    #modified BFS
    root = pixel_list[0]
    np.delete(pixel_list, 0)
    queue = collections.deque()
    queue.appendleft(root)
    queue_empty = False
    cluster_pixels = np.array()
    np.append(cluster_pixels,root)
    height,width = partial_part_table.shape
    while queue_empty != True:
        v = queue.pop()
        queue.remove(v)
        adj_list = adj_pixel_list(v[0],v[1], partial_part_table)
        for pixel in adj_list:
            #need to check if within bounds
            if pixel[0] != 0 and pixel[0] != height+1 and pixel[1] != 0 and pixel[1]!=width+1:
                #if color index is correct and not discovered
                if partial_part_table[pixel[0]][pixel[1]][0] == index and partial_part_table[pixel[0]][pixel[1]][1]==0:
                    #assign cluster in the part table, add to cluster list
                    partial_part_table[pixel[0]][pixel[1]][1]=curr_cluster
                    remaining_pixel_list = remove_from_list(remaining_pixel_list,pixel)
                    np.append(cluster_pixels,pixel)
                    queue.appendleft(pixel)
            
        try: 
            val = queue.pop()
            queue.append(val)
        except IndexError: 
            queue_empty = True
    return cluster_pixels
            
#img just needs to be an array that is the same size as the image:
    #part_tables/color arrays all work
#returns [top,bottom,left,right] where each element is either [row,col] or [-1,-1]
def adj_pixel_list(row, column, img):
    height,width = shape(img)
    adj_list = np.array()
    np.append(adj_list, [row-1,column])
    np.append(adj_list,[row+1,column])
    np.append(adj_list,[row,column-1])
    np.append(adj_list,[row,column+1])
    return adj_list

def remove_from_list(pixel_list,pixel):
    for index in np.arange(len(pixel_list)):
        if pixel_list[index] == pixel:
            return np.delete(pixel_list,index)
    
#returns an array [cluster pixel, reference pixel]
    #example: [[0,0],[1,1]]
#takes a part table and finds the closest reference to a pixel that is not labeled -1 or the edge
def find_closest_references(cluster_list, partial_part_table,height,width,index):
    min_dist = 32767
    #bfs on random pixel in cluster by building a tree with pixels as nodes and:
        #- the cluster pixel that led to the lowest distance to that pixel
        #- the distance from the closest cluster pixel
    #this modification of BFS updates the distance metadata on a node if shorter distance paths are found to that node
        #and only enqueues nodes in the form of [node, cluster_p,dist] if 
            #a) the pixel is not a cluster pixel and it has a shorter distance than the recorded pixel
            #b) the pixel is a cluster pixel, and so pixel and cluster_p are the same and dist is still MAX
    #final tree is represented as an array of nodes, where each node is represented by [curr_pixel, shortest_prev_node, index_of_next_nodes,dist_from_cluster,cluster_or_empty_or_dest, index in tree]
    
    dist = np.array((height+2,width+2), [32767,0]) #image represented as [distance from cluster, discovered_or_not]
    prev = np.array((height+2,width+2), [-1,-1]) #contains the coords of the previous pixel in the closest path towards a pixel
    root = cluster_list[0]
    queue = collections.deque()
    queue.appendleft(root)
    queue_empty=False
    while queue_empty != True:
        node = queue.pop()
        adj_list = adj_pixel_list(node[0],node[1],partial_part_table)
        for pixel in adj_list:
            
            modified_loc = np.array([pixel[0]+1,pixel[1]+1])
            
            if pixel[0] == -1 or pixel[1] == -1:
                pixel_index = -2
            else:
                pixel_index = partial_part_table[pixel[0]][pixel[1]][0]
                
            pixel_status = dist[modified_loc[0]][modified_loc[1]][1]
            
            new_dist = 32767
            node_dist = dist[node[0]][node[1]][0]
            if node_dist > 1000:
                 new_dist = 1
            else:
                 new_dist = node_dist + 1

            #if pixel hasn't been processed and is the same color as cluster
            if pixel_index == index and pixel_status == 0:
                #process and add to queue
                dist[modified_loc[0]][modified_loc[1]][1] = 1 #label as discovered
                #don't need parent ref
                #update distance to be max
                dist[modified_loc[0]][modified_loc[1]][0] = 32767
                #enqueue current pixel
                queue.appendleft(pixel)
                
                
            #might not be necessary, ignore repeats    
            #if pixel_index == index && pixel_status != 0:
                #check for unprocessed neighbors: if they exist, enqueue, else, end
                
            #if pixel is on an edge or other color and not processed:
            if pixel_index != -1 and pixel_status == 0:
                prev[modified_loc[0]][modified_loc[1]] = node
                dist[modified_loc[0]][modified_loc[1]][0] = new_dist
                dist[modified_loc[0]][modified_loc[1]][1] = 1
                
                #add to tree with dist of parent +1 (or just 1 if parent is a cluster node)
            
            #if pixel is on an edge or other color and already processed:
            if pixel_index != -1 and pixel_status != 0:
                #check existing node in the tree and see if the distance is lower; if so, update parent and end
                old_prev = prev[modified_loc[0]][modified_loc[1]]
                if dist[old_prev[0]][old_prev[1]] > new_dist:
                    dist[old_prev[0]][old_prev[1]] = new_dist
                    prev[modified_loc[0]][modified_loc[1]] = node
                
            # if an empty pixel:
            if pixel_index == -1:
                #check if part of previous path
                if prev[modified_loc[0]][modified_loc[1]][0] == -1 or prev[modified_loc[0]][modified_loc[1]][1]==-1:
                    prev[modified_loc[0]][modified_loc[1]]=node
                #check distance, update and enqueue if necessary
                if dist[modified_loc[0]][modified_loc[1]] > new_dist:
                    dist[modified_loc[0]][modified_loc[1]] = new_dist
                    prev[modified_loc[0]][modified_loc[1]]=node
                    queue.appendleft(pixel)
                
    #queue is now empty
    #traverse tree to find the shortest distance destination cluster, then traverse up parent by parent to find cluster pix
        #for all nodes in tree, find the lowest distance pixel that does not have an index of -1 
    min_dist = 32767
    min_loc = [-1,-1]
    for row in np.arange(height+2):
        for column in np.arange(width+2):
            if dist[row][column] < min_dist:
                min_dist = dist[row][column]
                min_loc[0]=row
                min_loc[1]=column
    cluster_loc = [-1,-1]
    curr_row = min_loc[0]
    curr_col = min_loc[1]
    while prev[curr_row][curr_col][0] != -1:
        curr_row = prev[curr_row][curr_col][0]
        curr_col = prev[curr_row][curr_col][1]
    cluster_loc[0] = curr_row
    cluster_loc[1] = curr_col
    return np.array([cluster_loc,min_loc])
    
    
```
